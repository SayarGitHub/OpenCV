# Reading Image

```python
import matplotlib.pyplot as plt
# Importing the OpenCV library 
import cv2 
# Reading the image using imread() function 
image = cv2.imread('road.jpg')

# cv2.IMREAD_COLOR: It specifies to load a color image. 
# Any transparency of image will be neglected. It is the default flag. 
# Alternatively, we can pass integer value 1 for this flag.

# cv2.IMREAD_GRAYSCALE: It specifies to load an image in grayscale mode. 
# Alternatively, we can pass integer value 0 for this flag.

# cv2.IMREAD_UNCHANGED: It specifies to load an image as such including alpha channel. 
# Alternatively, we can pass integer value -1 for this flag.

# Extracting the height and width of an image 
h, w = image.shape[:2] 
# Displaying the height and width 
print("Height = {}, Width = {}".format(h, w))

# Display image in a window after resizing
ims = cv2.resize(image, (1280, 720))  
cv2.imshow("Road", ims)
cv2.waitKey(0)

cv2.destroyAllWindows()

# Display inline
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

# Extracting the RGB values of a pixel

```python
# Extracting RGB values. 
# Here we have randomly chosen a pixel 
# by passing in 100, 100 for height and width. 
(B, G, R) = image[100, 100] # opencv uses BGR format

# Displaying the pixel values 
print("R = {}, G = {}, B = {}".format(R, G, B)) 

# We can also pass the channel to extract 
# the value for a specific channel 
B = image[100, 100, 0] 
print("B = {}".format(B))
```

# Extracting the Region of Interest (ROI)

```python
# We will calculate the region of interest 
# by slicing the pixels of the image 
roi = image[100 : 500, 200 : 700]
cv2.imshow("Cropped", roi)
cv2.waitKey(0)

cv2.destroyAllWindows() 

plt.imshow(cv2.cvtColor(roi,cv2.COLOR_BGR2RGB))
plt.show()
```

# Resizing the Image

```python
#Fixing distortion when resizing
# Calculating the ratio 
ratio = 800 / w 

# Creating a tuple containing width and height 
dim = (800, int(h * ratio)) 

# Resizing the image 
resize_aspect = cv2.resize(image, dim) 

plt.imshow(cv2.cvtColor(resize_aspect,cv2.COLOR_BGR2RGB))
plt.show()
```

# Rotating the Image

```python
# Calculating the center of the image 
center = (w // 2, h // 2) 

# Generating a rotation matrix 
matrix = cv2.getRotationMatrix2D(center, -45, 1.0)  #(point of rotation, angle, scale)

# Performing the affine transformation 
rotated = cv2.warpAffine(image, matrix, (w, h)) 

plt.imshow(cv2.cvtColor(rotated,cv2.COLOR_BGR2RGB))
plt.show()
```

# Drawing a Rectangle

```python
# We are copying the original image, 
# as it is an in-place operation. 
output = image.copy() 

# Using the rectangle() function to create a rectangle. 
rectangle = cv2.rectangle(output, (1500, 900), 
						(600, 400), (255, 0, 0), 5)

#rectangle takes following arguments-   

# Image
# Top-left corner co-ordinates
# Bottom-right corner co-ordinates
# Color (in BGR format)
# Line width

plt.imshow(cv2.cvtColor(output,cv2.COLOR_BGR2RGB))
plt.show()
```

# Displaying text

```python
# Copying the original image 
output = image.copy() 

# Adding the text using putText() function 
text = cv2.putText(output, 'OpenCV Demo', (500, 550), 
				cv2.FONT_HERSHEY_SIMPLEX, 4, (0, 255, 0), 5) 

# It takes in 7 arguments –

# Image
# Text to be displayed
# Bottom-left corner co-ordinates, from where the text should start
# Font
# Font size
# Color (BGR format)
# Line width

plt.imshow(cv2.cvtColor(output,cv2.COLOR_BGR2RGB))
plt.show()
```
